```python
df = pd.read_csv(r'C:\Users\User\OneDrive\Documents\Cardiohealth\bmi_pop_data.csv') 
```

```python
print(df.head())
```

```python
# Convert cardio_condition to boolean
df['cardio_condition'] = df['cardio_condition'].apply(lambda x: True if x == 1 else False)
```

```python
print(df.head())
```

```python
df_sorted_asc = df.sort_values('age_in_years')
```

```python
print(len(df))
```

```python
# Count the frequency of each BMI rank/category
bmigendergroup_counts = df.groupby(['bmi_class', 'cardio_condition', 'gender']).size().unstack(['cardio_condition', 'gender'])
bmigroup_counts =df.groupby(['bmi_class', 'cardio_condition']).size().unstack(fill_value=0) 
bmicategory_counts = df['bmi_class'].value_counts()
bmicategory_counts.columns = ['bmi_class', 'frequency']
```

```python

# Create the bar chart
plt.figure(figsize=(8, 6))
bmicategory_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of BMI Classes')
plt.xlabel('BMI Category')
plt.ylabel('Number of Participants')
plt.show()

# Display the plot
plt.tight_layout()
plt.show()

# Save the plot as a PNG image
plt.savefig('BMIcategory_distribution_bar_chart.png')




```

```python
# Create the bar chart
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.35

x = np.arange(len(bmigroup_counts.index))  # Corrected: Use bmi_counts.index for x-axis positions

rects1 = ax.bar(x - width/2, bmigroup_counts[False], width, label='No CVD', color='skyblue')
rects2 = ax.bar(x + width/2, bmigroup_counts[True], width, label='CVD', color='navy')

# Set chart title and labels
ax.set_title('Distribution of BMI Classes by CVD Status')
ax.set_xlabel('BMI Category')
ax.set_ylabel('Number of Participants')
ax.set_xticks(x)
ax.set_xticklabels(bmigroup_counts.index)
ax.legend()

# Add data labels on top of the bars
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.tight_layout()
plt.show()
```

```python
# Create the bar chart
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.2 

# Calculate x-positions for each group
x = np.arange(len(bmigendergroup_counts.index))
x_male_no_cvd = x - 1.5*width
x_male_cvd = x - 0.5*width
x_female_no_cvd = x + 0.5*width
x_female_cvd = x + 1.5*width

# Create bars for each group
rects1 = ax.bar(x_male_no_cvd, bmigendergroup_counts[(False, 'male')], width, label='Male - No CVD', color='skyblue')
rects2 = ax.bar(x_male_cvd, bmigendergroup_counts[(True, 'male')], width, label='Male - CVD', color='navy')
rects3 = ax.bar(x_female_no_cvd, bmigendergroup_counts[(False, 'female')], width, label='Female - No CVD', color='lightcoral')
rects4 = ax.bar(x_female_cvd, bmigendergroup_counts[(True, 'female')], width, label='Female - CVD', color='maroon')

# Set chart title and labels
ax.set_title('Distribution of BMI Classes by CVD Status and Gender')
ax.set_xlabel('BMI Category')
ax.set_ylabel('Number of Participants')
ax.set_xticks(x)
ax.set_xticklabels(bmigendergroup_counts.index)
ax.legend()

# Add data labels on top of the bars
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

plt.tight_layout()
plt.show()
```
